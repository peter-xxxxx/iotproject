```python
# Again we import the necessary socket python module
import socket
import time
# Here we define the UDP IP address as well as the port number that we have 
# already defined in the client python script.
UDP_IP_ADDRESS = "192.168.0.10"
UDP_PORT_NO = 6001

# declare our serverSocket upon which
# we will be listening for UDP messages
serverSock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
# One difference is that we will have to bind our declared IP address
# and port number to our newly declared serverSock
serverSock.bind((UDP_IP_ADDRESS, UDP_PORT_NO))

data = []
while True:
    message, addr = serverSock.recvfrom(1024)
    
    data.append(message.decode("utf-8"))
    
    if len(data)>=100:
        break
    
    

```

```python
len(data)
```

```python
ax = []
ay = []
az = []
gx = []
gy = []
gz = []

for da in data:
    ax.append(float(da.split(' ')[0]))
    ay.append(float(da.split(' ')[1]))
    az.append(float(da.split(' ')[3]))
    gx.append(float(da.split(' ')[3]))
    gy.append(float(da.split(' ')[4]))
    gz.append(float(da.split(' ')[5]))
```

```python
gyro = []
accm = []
```

```python
gyro.append(gx)
gyro.append(gy)
gyro.append(gz)
accm.append(ax)
accm.append(ay)
accm.append(az)

print(len(gyro))
print(len(accm))
```

```python
import matplotlib.pyplot as plt
xaxis_n = 25
xaxis = list(range(xaxis_n))

plt.figure(figsize=(20,10))

plt.subplot(321)
plt.plot(xaxis,gyro[0][:xaxis_n],xaxis,gyro[3][:xaxis_n],xaxis,gyro[6][:xaxis_n])
plt.title('gx')

plt.subplot(322)
plt.plot(xaxis,accm[0][:xaxis_n],xaxis,accm[3][:xaxis_n],xaxis,accm[6][:xaxis_n])
plt.title('ax')

plt.subplot(323)
plt.plot(xaxis,gyro[1][:xaxis_n],xaxis,gyro[4][:xaxis_n],xaxis,gyro[7][:xaxis_n])
plt.title('gy')

plt.subplot(324)
plt.plot(xaxis,accm[1][:xaxis_n],xaxis,accm[4][:xaxis_n],xaxis,accm[7][:xaxis_n])
plt.title('ay')

plt.subplot(325)
plt.plot(xaxis,gyro[2][:xaxis_n],xaxis,gyro[5][:xaxis_n],xaxis,gyro[8][:xaxis_n])
plt.title('gz')

plt.subplot(326)
plt.plot(xaxis,accm[2][:xaxis_n],xaxis,accm[5][:xaxis_n],xaxis,accm[8][:xaxis_n])
plt.title('az')

plt.show()

```
